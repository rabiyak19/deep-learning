```python
import pandas as pd
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Embedding, Dense, Flatten
from sklearn.model_selection import train_test_split
```

```python
messages_data = pd.read_csv('C:/Users/Bilal/Desktop/deep_learning/dataset/spam_ham_classification.csv', encoding='ISO-8859-1')
```

```python
labels = messages_data['v1'].apply(lambda x: x.lower() == 'spam').astype('int')
```

```python
labels
```

```python
messages = messages_data['v2'].values
messages
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(messages)
messages_token = tokenizer.texts_to_sequences(messages)
```

```python
print(messages_token[1], messages[1])
```

```python
len(messages_token[0]), len(messages_token[1])
```

```python
tokenizer.word_index
```

```python
tokenizer.word_index.keys()
```

```python
messages_token = pad_sequences(messages_token, maxlen=100)
```

```python
len(messages_token[0]), len(messages_token[1])
```

```python
labels.value_counts()
```

```python
#try downgrading to see if it improves acc or not
```

```python
X_train, X_test, y_train, y_test = train_test_split(messages_token, labels, test_size=0.2, random_state=67)
```

```python
model = Sequential()
model.add(Embedding(len(tokenizer.word_index.keys()) + 1, 64))
model.add(LSTM(64))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=5, validation_split=0.2, batch_size=30)
```

```python
result = model.evaluate(X_test, y_test)
```
